```python
Q1. what are the five key concepts of object-oriented programming(OOP)?
Answer: bject-oriented programming (OOP) is a paradigm centered around objects and classes, offering a way to structure and organize code. Here are the five key concepts:

Classes and Objects:

Class: A blueprint or prototype defining attributes and behaviors (methods) that objects created from the class can have.

Object: An instance of a class.
-----------------------------------------------------------------------------------------------------------------------------------------
Encapsulation:

Bundling data (attributes) and methods (functions) that operate on the data into a single unit or class.

Restricting direct access to some of an object's components, which is a means of preventing accidental interference and misuse of the data.
------------------------------------------------------------------------------------------------------------------------------------------------
Inheritance:

A mechanism to create a new class from an existing class.

The new class (child or subclass) inherits attributes and methods of the existing class (parent or superclass),
allowing for code reuse and the creation of a more specific implementation.
--------------------------------------------------------------------------------------------------------------------------------------------------
Polymorphism:

The ability of different objects to respond to the same method call in different ways.

Achieved through method overriding (a subclass provides a specific implementation of a method already defined in its superclass) or
method overloading (multiple methods have the same name but different parameters in the same class).
----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
Abstraction:

The concept of hiding the complex implementation details and showing only the essential features of the object.

Helps in reducing programming complexity and effort.

OOP makes your code more modular, flexible, and easier to maintain.
```

```python
Q5. what are the three types of access modifiers in python?how are they donated?

Answer:n Python, there are three primary types of access modifiers that control the visibility of class members (attributes and methods). These are:

1. Public
Syntax: No special notation is required.

Visibility: Accessible from anywhere (inside or outside the class).
-------------------------------------------------------------------------------------------------------------------------------------------------
Example:

python
class MyClass:
    def __init__(self):
        self.public_variable = 10  # Public variable

    def public_method(self):
        print("This is a public method")

obj = MyClass()
print(obj.public_variable)  # Accessing public variable
obj.public_method()  # Calling public method
-----------------------------------------------------------------------------------------------------------------------------------------------
Protected
Syntax: Prefixed with a single underscore (_).

Visibility: Should not be accessed outside the class or its subclasses. It's more of a convention rather than enforced by Python.

Example:

python
class MyClass:
    def __init__(self):
        self._protected_variable = 20  # Protected variable

    def _protected_method(self):
        print("This is a protected method")

obj = MyClass()
print(obj._protected_variable)  # Accessing protected variable (not recommended)
obj._protected_method()  # Calling protected method (not recommended)
------------------------------------------------------------------------------------------------------------------------------------------------
Private
Syntax: Prefixed with double underscores (__).

Visibility: Not accessible directly from outside the class. Name mangling is used to make it harder to access these attributes and methods.

Example:

python
class MyClass:
    def __init__(self):
        self.__private_variable = 30  # Private variable

    def __private_method(self):
        print("This is a private method")

    def access_private(self):
        # Method to access private attributes
        print(self.__private_variable)
        self.__private_method()

obj = MyClass()
# Trying to access private variable and method directly will result in an error
# print(obj.__private_variable)  # AttributeError
# obj.__private_method()  # AttributeError

# Accessing private variable and method through a public method
obj.access_private()
-------------------------------------------------------------------------------------------------------------------------------------------------
Summary
Public: No special notation, accessible from anywhere.

Protected: Single underscore (_), convention to indicate protected access, intended for use within the class and subclasses.

Private: Double underscore (__), name mangling to make it harder to access directly from outside the class.

These access modifiers help in defining the intended scope of class members and provide a way to encapsulate the data and methods within a class, 
promoting better design and code readability. 🚀
```

```python
Q6.Describe the five types of inheritance in python.provide a simple example of inheritance.

Answer;Sure, let's dive into the five types of inheritance in Python and provide a simple example:
-----------------------------------------------------------------------------------------------------------------------------------------
1. Single Inheritance
Single inheritance allows a derived class to inherit from a single base class.

Example:

python
class Vehicle:
    def info(self):
        print("This is a vehicle.")

class Car(Vehicle):
    def car_info(self):
        print("This is a car.")

my_car = Car()
my_car.info()      # Inherited from Vehicle class
my_car.car_info()  # Car class method
----------------------------------------------------------------------------------------------------------------------------------------
2. Multiple Inheritance
Multiple inheritance allows a derived class to inherit from more than one base class.

Example:

python
class Engine:
    def engine_info(self):
        print("This is an engine.")

class Wheels:
    def wheels_info(self):
        print("These are wheels.")

class Car(Engine, Wheels):
    def car_info(self):
        print("This is a car.")

my_car = Car()
my_car.engine_info()  # Inherited from Engine class
my_car.wheels_info()  # Inherited from Wheels class
my_car.car_info()     # Car class method
------------------------------------------------------------------------------------------------------------------------------------------------
3. Multilevel Inheritance
Multilevel inheritance is a chain of inheritance where a class is derived from another derived class.

Example:

python
class Vehicle:
    def vehicle_info(self):
        print("This is a vehicle.")

class Car(Vehicle):
    def car_info(self):
        print("This is a car.")

class SportsCar(Car):
    def sports_car_info(self):
        print("This is a sports car.")

my_sportscar = SportsCar()
my_sportscar.vehicle_info()   # Inherited from Vehicle class
my_sportscar.car_info()       # Inherited from Car class
my_sportscar.sports_car_info() # SportsCar class method
----------------------------------------------------------------------------------------------------------------------------------------------------
4. Hierarchical Inheritance
In hierarchical inheritance, multiple derived classes inherit from a single base class.

Example:

python
class Vehicle:
    def vehicle_info(self):
        print("This is a vehicle.")

class Car(Vehicle):
    def car_info(self):
        print("This is a car.")

class Truck(Vehicle):
    def truck_info(self):
        print("This is a truck.")

my_car = Car()
my_truck = Truck()

my_car.vehicle_info()    # Inherited from Vehicle class
my_car.car_info()        # Car class method

my_truck.vehicle_info()  # Inherited from Vehicle class
my_truck.truck_info()    # Truck class method
-----------------------------------------------------------------------------------------------------------------------------------------------------
5. Hybrid Inheritance
Hybrid inheritance is a combination of two or more types of inheritance. It aims to exploit the advantages of various types of inheritance.

Example:

python
class Vehicle:
    def vehicle_info(self):
        print("This is a vehicle.")

class Engine:
    def engine_info(self):
        print("This is an engine.")

class Car(Vehicle, Engine):
    def car_info(self):
        print("This is a car.")

class Truck(Vehicle):
    def truck_info(self):
        print("This is a truck.")

my_car = Car()
my_truck = Truck()

my_car.vehicle_info()    # Inherited from Vehicle class
my_car.engine_info()     # Inherited from Engine class
my_car.car_info()        # Car class method

my_truck.vehicle_info()  # Inherited from Vehicle class
my_truck.truck_info()    # Truck class method
These examples demonstrate the different types of inheritance in Python. 
```

```python
Q7. What is method resolution order(MRO) in python? hoe can you retrieve it programmatically?

Answer; The Method Resolution Order (MRO) in Python determines the order in which base classes are searched when executing a method.
This is especially important in multiple inheritance scenarios, where a class is derived from more than one base class.
Python uses the C3 linearization algorithm to determine the MRO.
------------------------------------------------------------------------------------------------------------------------------------------------------
How to Retrieve MRO Programmatically
You can retrieve the MRO of a class using the mro() method or the __mro__ attribute. Here’s an example:

python
class A:
    pass

class B(A):
    pass

class C(A):
    pass

class D(B, C):
    pass

# Using the mro() method
print(D.mro())

# Using the __mro__ attribute
print(D.__mro__)

Both of these will output the Method Resolution Order for the D class:

[<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]
                                                     
Explanation
The output shows the order in which classes will be searched for a method, starting from D and following the hierarchy defined by the inheritance.

MRO ensures that the method resolution is predictable and consistent, particularly when dealing with complex inheritance hierarchies.
```

```python
Q10. Implement encapsulation in a 'bankaccount' class with private attribute for 'balance' and 'account_number'. 
include for deposit ,withdraw, and balance inquiry.

Answer:Sure! Encapsulation in object-oriented programming is the concept of restricting access to certain attribute
and methods to protect the integrity of the data.
This is often done using private attributes and providing public methods to access and modify those attributes.
----------------------------------------------------------------------------------------------------------------------------------------
Here's an example of a BankAccount class with encapsulation implemented:
       
class BankAccount:
    def __init__(self, account_number, initial_balance=0):
        self.__account_number = account_number  # Private attribute
        self.__balance = initial_balance  # Private attribute

    # Public method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return True
        else:
            print("Deposit amount must be positive.")
            return False

    # Public method to withdraw money
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return True
        else:
            print("Invalid withdrawal amount.")
            return False

    # Public method to inquire balance
    def get_balance(self):
        return self.__balance

    # Public method to get the account number
    def get_account_number(self):
        return self.__account_number

# Example usage
account = BankAccount("1234567890", 1000)
print(f"Account Number: {account.get_account_number()}")
print(f"Initial Balance: {account.get_balance()}")

account.deposit(500)
print(f"Balance after deposit: {account.get_balance()}")

account.withdraw(300)
print(f"Balance after withdrawal: {account.get_balance()}")
---------------------------------------------------------------------------------------------------------------------------------------------
Explanation
Private Attributes: __account_number and __balance are prefixed with double underscores to indicate they are private attributes. 
This means they cannot be accessed directly from outside the class.

Public Methods: Methods like deposit, withdraw, get_balance, and get_account_number provide controlled access to these private attributes.

Validation: The deposit and withdraw methods include validation to ensure that the operations are valid 
(e.g., positive deposit amount, sufficient balance for withdrawal).
-------------------------------------------------------------------------------------------------------------------------------------------------
Benefits
Data Protection: Encapsulation ensures that the internal state of the object is protected from unauthorized access and modification.

Controlled Access: Public methods provide a controlled way to access and modify the private attributes, ensuring that any necessary validation or business logic is applied.

This approach helps maintain the integrity and consistency of your data, making your code more robust and easier to maintain!
```

```python
Q12. create a decorator that measures and prints the execution time of a function.

Answer:ure! Here's a Python decorator that measures and prints the execution time of a function.
Decorators are a powerful way to add functionality to your functions in a clean and readable manner.
----------------------------------------------------------------------------------------------------------------------------------------------
Decorator Code
python
import time

def measure_time(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"Function '{func.__name__}' executed in {execution_time:.4f} seconds")
        return result
    return wrapper

# Example usage
@measure_time
def slow_function(seconds):
    time.sleep(seconds)
    return "Finished"

# Running the example
print(slow_function(2))
-------------------------------------------------------------------------------------------------------------------------------------------------
Explanation
measure_time: This is the decorator function. It takes a function func as an argument.

wrapper: This inner function wraps around func to measure its execution time. It takes any arguments and keyword arguments that the original function might take.

Timing: start_time and end_time are captured using time.time(), and the difference gives the execution time.

Print Execution Time: The execution time is printed in a human-readable format.

Return Result: The result of the original function is returned unchanged.
-------------------------------------------------------------------------------------------------------------------------------------------------
Benefits
Performance Monitoring: Quickly measure and print how long functions take to execute.

Ease of Use: Simply add @measure_time above any function to start monitoring its performance.
Ease of Use: Simply add @measure_time above any function to start monitoring its performance.
```

```python
Q13. Explain the concept of the diamond problem in multiple inheritance. How does python resolve it?

Answer: he diamond problem occurs in multiple inheritance when a class inherits from two classes that have a common base class. 
The issue arises when methods or attributes from the common base class are inherited by the child class, creating ambiguity.
--------------------------------------------------------------------------------------------------------------------------------------------
Example of the Diamond Problem
Consider the following class hierarchy:

python
class A:
    def method(self):
        print("Method from class A")

class B(A):
    def method(self):
        print("Method from class B")

class C(A):
    def method(self):
        print("Method from class C")
 class D(B, C):
    pass

d = D()
d.method()
-------------------------------------------------------------------------------------------------------------------------------------------------
In this example:

Class A is the common base class.

Class B and Class C both inherit from Class A and override the method.

Class D inherits from both Class B and Class C.

The ambiguity is which method should Class D inherit and execute, since both Class B and Class C override the method of Class A.

Python's Resolution: Method Resolution Order (MRO)
Python resolves the diamond problem using the Method Resolution Order (MRO), which is based on the C3 linearization algorithm. This algorithm ensures a consistent order in which classes are to be accessed, avoiding ambiguity.

To see the MRO, you can use the __mro__ attribute or the mro() method. In our example:

print(D.__mro__)
or

print(D.mro())
This will output:

(<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)
MRO in Action
When d.method() is called, Python follows this order:

D: It doesn't have method, so it moves on to:

B: It finds method here and stops searching.

Thus, d.method() will print:

Method from class B
------------------------------------------------------------------------------------------------------------------------------------------------------    
Summary
Diamond Problem: Ambiguity in multiple inheritance when methods/attributes from a common base class are inherited.

MRO: Python resolves this using the C3 linearization algorithm, providing a consistent and unambiguous method resolution order.

This mechanism makes Python's handling of multiple inheritance robust and predictable, ensuring that the correct methods are called without ambiguity
```

```python
Q15. Implement a static method in a class that checks if a given year is a leap year.

Answer: Sure! Here's how you can implement a static method in a class to check if a given year is a leap year:
---------------------------------------------------------------------------------------------------------------------------------------------------
Implementation
python
class YearUtils:
    @staticmethod
    def is_leap_year(year):
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            return True
        else:
            return False

# Example usage
print(YearUtils.is_leap_year(2020))  # Output: True
print(YearUtils.is_leap_year(2021))  # Output: False
print(YearUtils.is_leap_year(1900))  # Output: False
print(YearUtils.is_leap_year(2000))  # Output: True
--------------------------------------------------------------------------------------------------------------------------------------------------
Explanation
Static Method: is_leap_year is defined as a static method using the @staticmethod decorator. This means it can be called on the class itself, without needing an instance.

Leap Year Check: The method checks if a year is divisible by 4, but not by 100, unless it is also divisible by 400:

A year is a leap year if it is divisible by 4.

However, if it is divisible by 100, it is not a leap year unless it is also divisible by 400.

Example Usage: You can call this method directly on the class YearUtils, passing the year you want to check as an argument.

Benefits
Utility Function: Static methods are useful for creating utility functions that are relevant to the class but don't depend on instance-specific data.

Reusability: The is_leap_year method can be used without creating an instance of YearUtils, 
making it more convenient to use in different parts of your code.
```
